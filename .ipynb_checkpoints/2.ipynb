```python
#Importar as bibliotecas para manipular os dados

import pandas as pd
import numpy as np
df = pd.read_csv('airlines_flights_data.csv')
```

```python
df.info()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['price'] = df['price'].astype('float64')
print(df.info())
```

```python
mapeamento_nomes = {
    'Air_India': 'Air India',
    'GO_FIRST': 'Go First'
}
df['airline'] = df['airline'].replace(mapeamento_nomes, regex = False)

mapeamento_horarios = {
    'Early_Morning': 'Early Morning',
    'Late_Night': 'Late Nigth'
}
df['departure_time'] = df['departure_time'].replace(mapeamento_horarios, regex = False)
```

*Análise exploratória*

```python
#Importar as bibliotecas para manipular o gráfico e definir os parâmetros principais

import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

*Quais as companhias aéreas são as mais populares?*

```python
df['airline'].value_counts()
```

```python
plt.figure(figsize=(8,5)) #ajustando tamanho da figura / largura x altura
sns.countplot(data=df, x='airline', order=df['airline'].value_counts().index)
#data=df: a função vai usar o dataframe df
#x='airline' = as colunas do gráfico vão ser retiradas da tabela airline, o seaborn conta automaticamente a freq
#order=df['airline'].value_counts().index: Garante que as barras sejam ordenadas

#Adicionando um título e rótulos para os eixos
plt.title('Número de Voos por Companhia Aérea')
plt.xlabel('Companhia Aérea', labelpad = 20)
plt.ylabel('Número de Voos', labelpad = 20)

#Gire os rótulos do eixo x se eles estiverem sobrepostos
plt.xticks(rotation=0, ha='right')

plt.show()
```

*Distribuição dos preços*

```python
# Crie um histograma para a coluna 'price'
sns.histplot(data=df, x='price', bins=50, kde=True)

# Adicione um título e rótulos para os eixos para facilitar a leitura
plt.title('Distribuição de Preços das Passagens Aéreas')
plt.xlabel('Preço', labelpad = 20)
plt.ylabel('Frequência (Número de Voos)', labelpad = 20)

# Exiba o gráfico
plt.show()
```

*Mapa de Calor*

```python
#Selecionando as colunas
colunas_numericas = df.select_dtypes(include = ['int64', 'float64']).columns

corrmat =df[colunas_numericas].corr()

plt.figure(figsize=(8,5))

mask = np.triu(np.ones_like(corrmat, dtype = bool))
sns.heatmap(corrmat, vmax = .8, mask=mask, square = True, annot = True)
plt.show()
```

```python
df['source_city'].value_counts()
df['departure_time'].value_counts()

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='source_city', order=df['source_city'].value_counts().index)

plt.title('Distribuição de Voos por Cidade de Origem')
plt.xlabel('Cidade de Origem', labelpad = 20)
plt.ylabel('Número de Voos', labelpad = 20)
plt.xticks(rotation=0, ha='right')

plt.show()

# Gráfico de barras para a contagem de voos por cidade de destino
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='destination_city', order=df['destination_city'].value_counts().index)

plt.title('Distribuição de Voos por Cidade de Destino')
plt.xlabel('Cidade de Destino', labelpad = 20)
plt.ylabel('Número de Voos', labelpad =20)

plt.xticks(rotation=0, ha='right')
plt.show()


# Gráfico de barras para a contagem de voos por horário de partida
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='departure_time', order=df['departure_time'].value_counts().index)

plt.title('Distribuição de Voos por Horário de Partida')
plt.xlabel('Horário de Partida', labelpad = 20)
plt.ylabel('Número de Voos', labelpad = 20)
plt.show()
```

*Relação de Preços por Classe de Voo*

```python
#Agrupondo os dados pela coluna class e calculando media e mediana
df.groupby('class')['price'].mean()

plt.figure(figsize = (8, 5))
sns.boxplot(data=df, x='class', y='price')

plt.title('Relação de Preços por Classe de Voo')
plt.xlabel('Classe de Voo', labelpad =20)
plt.ylabel('Preço', labelpad =20)
plt.show()
```

*Relação entre preço e número de dais restantes para o voo*

```python
plt.figure(figsize = (8,5))

sns.scatterplot(data=df, x='days_left', y='price', alpha=0.5, s=20)

plt.title('Relação entre Preço e Dias Restantes para o Voo')
plt.xlabel('Dias Restantes para o Voo', labelpad =20)
plt.ylabel('Preço da Passagem', labelpad =20)
plt.show()
```

*A duração do voo afeta o preço da passagem?*

```python
plt.figure(figsize = (8,5))

sns.scatterplot(data=df, x='duration', y='price', alpha=0.5, s=20)

plt.title('Relação entre Preço e duração do Voo')
plt.xlabel('Duração do Voo (hrs)', labelpad =20)
plt.ylabel('Preço da Passagem (₹)', labelpad =20)
plt.show()
```

```python
# Primeiro, calcule o preço médio por companhia aérea e ordene
media_precos_por_companhia = df.groupby('airline')['price'].mean().sort_values(ascending=False)

# Crie um gráfico de barras na horizontal
plt.figure(figsize=(8, 5))
sns.barplot(x=media_precos_por_companhia.values, y=media_precos_por_companhia.index)

# Adicione um título e rótulos
plt.title('Preço Médio por Companhia Aérea', pad=20) # Adicionei uma margem
plt.xlabel('Preço Médio', labelpad=20)
plt.ylabel('Companhia Aérea', labelpad=20)

# Exiba o gráfico
plt.show()
```

```python
#Variável para armazenar o df filtrado
voos_filtrados = df[
    (df['airline'] == 'Vistara') &
    (df['source_city'] == 'Delhi') &
    (df['destination_city'] == 'Hyderabad')&
    (df['class'] == 'Business')]

#Calcular o preço médio do df filtrado
preco_medio = voos_filtrados['price'].mean()

print(f"O preço médio da companhia Vistara para um voo de Delhi para Hyderabad em classe Business é: {preco_medio:.2f}₹", )
```
